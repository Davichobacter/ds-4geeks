```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/job_sector_wage.csv"
```

```python
df = pd.read_csv(url)
```

```python
from tableone import TableOne
```

```python
df.describe()
```

```python
df['sector'].value_counts()
```

```python
df['job'].value_counts()
```

```python
df.groupby(by=['job'])[['consumption', 'wage', 'age']].agg(['mean', 'std'])
```

```python
df.groupby(by=['sector'])[['consumption', 'wage', 'age']].agg(['mean', 'std'])
```

```python
TableOne(df,
         nonnormal = ['consumption'],
         categorical=['sector'],
         groupby='job',
         pval=True,
         htest_name=True)
```
