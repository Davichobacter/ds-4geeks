The gravity model relate the amount of commerce with the size of countries, its
distance an its international trade.

Using the datasets:
- Locations
- Importations
- Exportations
- Gross domestic products
- Country names

The following work need to establish empirical evidence in Colombia, could
test any model to support or not the hypothesis, for instance:
$$ y_{ij} = β_0 + β_1 gpd_j + β_2 dist_{ij} $$
Where the $ y_{ij} $ is the amount of international trade (importations+exportations)
between the $ i $ country with the $ j $ country, $ gdp_j $ is the domestic product $ j $ and
$ dist_{ij} $ could be a measure of distance among the countries $ i $ and $ j $.
